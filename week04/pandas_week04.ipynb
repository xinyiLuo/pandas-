# 数据框重塑
##  Pandas的stack实现数据透视

stack()是将原来的列索引转成了最内层的行索引，这里是多层次索引，其中AB索引对应第三层，即最内层索引。


![](http://localhost:8888/files/111.png)
表格在行列方向上均有索引（类似于DataFrame），花括号结构只有“列方向”上的索引（类似于层次化的Series），结构更加偏向于堆叠（Series-stack，方便记忆）。stack函数会将数据从”表格结构“变成”花括号结构“，即将其行索引变成列索引

```python

df_独角兽_创始人_准备 =df[["企业名称","掌门人/创始人"]]
df_独角兽_创始人_准备.index.name="序号"
df_独角兽_创始人 =pd.merge(df_独角兽_创始人_准备.drop('掌门人/创始人', axis=1),\
                          df['掌门人/创始人'].str.split(',', expand=True)\
                                            .stack().reset_index(level=1,drop=True)
                                            .rename('掌门人/创始人'),\
                          on="序号")

df_独角兽_投资机构_准备 = df[["企业名称","部分投资机构"]]
df_独角兽_投资机构_准备.index.name="序号"
df_独角兽_投资机构 = pd.merge(df_独角兽_投资机构_准备.drop('部分投资机构', axis=1),\
                             df['部分投资机构'].str.split('、', expand=True)\
                                             .stack().reset_index(level=1, drop=True)
                                             .rename('部分投资机构'),\
                             on="序号")


display(df_独角兽_创始人)
display(df_独角兽_投资机构)
```

# 数据框聚合(merge)

## merge进行合并

* 数据会被复制
* 数目以多的一边为准

### merge的参数
on：列名，join用来对齐的那一列的名字，用到这个参数的时候一定要保证左表和右表用来对齐的那一列都有相同的列名。

left_on：左表对齐的列，可以是列名，也可以是和dataframe同样长度的arrays。

right_on：右表对齐的列，可以是列名，也可以是和dataframe同样长度的arrays。

left_index/ right_index: 如果是True的haunted以index作为对齐的key

how：数据融合的方法。

sort：根据dataframe合并的keys按字典顺序排序，默认是，如果置false可以提高表现。

### 企业名称对应多个投资公司
* 首先把部分机构和所有企业名称取出来。把部分投资机构取出来split列表 字典企业名称 contains循环遍历每个投资方
* 找某一投资方投资的所有企业 建字典contains做一个字典 企业名称作为values
* 循环每个sublist取出来的item就是存到投资机构里面的值
* .join把列表变成字符串用，隔开

* Merge两个表格进行合并，用，隔开 前后是两个表格，按照序号的索引
* \换下一行
* stack按照序号
* reset-index删掉第二级索引，用字典形式把返回的series（一维只有值和序号）变成dateframe(二维)

```python
# C-1 直接给完整代码：
df_投资方_企业名称 = df[["部分投资机构","企业名称"]]
所有投资机构 = [item for sublist in df_投资方_企业名称["部分投资机构"].str.split('[、,,]') for item in sublist]

touzi_dict = {}
for item in 所有投资机构:
    df_投资方_item = df_投资方_企业名称[df_投资方_企业名称["部分投资机构"].str.contains(item)]
    touzi_dict[item] =','.join([i for i in df_投资方_item["企业名称"].values])

投资方_独角兽企业_Series = pd.Series(touzi_dict)

df_投资方_独角兽企业 = pd.DataFrame({"投资方":投资方_独角兽企业_Series.index,"独角兽企业":投资方_独角兽企业_Series.values})
df_投资方_独角兽企业.index.name="序号"
df_投资方_独角兽企业 =pd.merge(df_投资方_独角兽企业.drop('独角兽企业', axis=1),\
                     df_投资方_独角兽企业['独角兽企业'].str.split(',', expand=True).stack().reset_index(level=1,drop=True).rename('独角兽企业'),\
                     on="序号")
df_投资方_独角兽企业
```
